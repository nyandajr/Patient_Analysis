```python
import pandas as pd

# Paths to the directories
inpatient_dir = 'R:/Patient_Analysis/Inpatient/'
outpatient_dir = 'R:/Patient_Analysis/Outpatient/'

# Load the datasets
inpatient_data = {
    2018: pd.read_csv(inpatient_dir + 'IN_WL 2018.csv'),
    2019: pd.read_csv(inpatient_dir + 'IN_WL 2019.csv'),
    2020: pd.read_csv(inpatient_dir + 'IN_WL 2020.csv'),
    2021: pd.read_csv(inpatient_dir + 'IN_WL 2021.csv')
}

outpatient_data = {
    2018: pd.read_csv(outpatient_dir + 'Op_WL 2018.csv'),
    2019: pd.read_csv(outpatient_dir + 'Op_WL 2019.csv'),
    2020: pd.read_csv(outpatient_dir + 'Op_WL 2020.csv'),
    2021: pd.read_csv(outpatient_dir + 'Op_WL 2021.csv')
}
```

```python
# Function to inspect a dataset
def inspect_data(data):
    inspection_report = {
        'First 5 Rows': data.head(),
        'Data Types': data.dtypes,
        'Missing Values': data.isnull().sum(),
        'Summary Statistics': data.describe(include='all')
    }
    return inspection_report

# Inspecting one of the inpatient datasets as an example
inspect_inpatient_2018 = inspect_data(inpatient_data[2018])
inspect_inpatient_2019 = inspect_data(inpatient_data[2019])
inspect_inpatient_2020 = inspect_data(inpatient_data[2020])
inspect_inpatient_2021 = inspect_data(inpatient_data[2021])

# Print the inspection report for inpatient 2018 data
print(inspect_inpatient_2018)
print(inspect_inpatient_2019)
print(inspect_inpatient_2020)
print(inspect_inpatient_2021)


```

```python
# Data cleaning function
def clean_data(data):
    # Dropping the 'Unnamed: 8' column if it exists
    if 'Unnamed: 8' in data.columns:
        data = data.drop(columns=['Unnamed: 8'])
    
    # Convert 'Archive_Date' to datetime format
    data['Archive_Date'] = pd.to_datetime(data['Archive_Date'], dayfirst=True)

    return data

# Cleaning each dataset
for year in inpatient_data:
    inpatient_data[year] = clean_data(inpatient_data[year])

for year in outpatient_data:
    outpatient_data[year] = clean_data(outpatient_data[year])
```

```python
import matplotlib.pyplot as plt

# Function to calculate yearly totals
def calculate_yearly_totals(data_dict):
    yearly_totals = {}
    for year, data in data_dict.items():
        total = data['Total'].sum()
        yearly_totals[year] = total
    return yearly_totals

# Calculating totals for inpatient and outpatient data
inpatient_totals = calculate_yearly_totals(inpatient_data)
outpatient_totals = calculate_yearly_totals(outpatient_data)

# Plotting the trends
plt.figure(figsize=(10, 6))
plt.plot(list(inpatient_totals.keys()), list(inpatient_totals.values()), label='Inpatient', marker='o')
plt.plot(list(outpatient_totals.keys()), list(outpatient_totals.values()), label='Outpatient', marker='o')
plt.title('Yearly Patient Totals: Inpatient vs Outpatient')
plt.xlabel('Year')
plt.ylabel('Total Patients')
plt.xticks(list(inpatient_totals.keys()))
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Function to analyze average waiting times with a column check
def analyze_average_waiting_times(data_dict):
    average_waiting_times = {}
    for year, data in data_dict.items():
        # Check if 'Specialty_Name' column exists
        if 'Specialty_Name' in data.columns:
            # Grouping by Specialty and Time Bands, then calculating the average 'Total'
            grouped_data = data.groupby(['Specialty_Name', 'Time_Bands'])['Total'].mean().unstack()
            average_waiting_times[year] = grouped_data
        else:
            print(f"Column 'Specialty_Name' not found in the dataset for the year {year}")
    return average_waiting_times

# Analyzing average waiting times for inpatient and outpatient data
inpatient_average_waiting_times = analyze_average_waiting_times(inpatient_data)
outpatient_average_waiting_times = analyze_average_waiting_times(outpatient_data)


print("Inpatient Average Waiting Times for 2018:")
print(inpatient_average_waiting_times[2018].head())  
```

```python
# Function to check for the existence of 'Specialty_Name' column in each dataset
def check_specialty_name_column(data_dict, data_type):
    for year, data in data_dict.items():
        if 'Specialty_Name' in data.columns:
            print(f"'Specialty_Name' column exists in the {data_type} dataset for the year {year}")
        else:
            print(f"'Specialty_Name' column NOT found in the {data_type} dataset for the year {year}")

# Checking for inpatient data
print("Checking Inpatient Data:")
check_specialty_name_column(inpatient_data, "inpatient")

# Checking for outpatient data
print("\nChecking Outpatient Data:")
check_specialty_name_column(outpatient_data, "outpatient")
```

```python
# Function to analyze average waiting times with adjusted column names
def analyze_average_waiting_times(data_dict, is_outpatient=False):
    average_waiting_times = {}
    specialty_col = 'Speciality' if is_outpatient else 'Specialty_Name'

    for year, data in data_dict.items():
        # Grouping by Specialty and Time Bands, then calculating the average 'Total'
        grouped_data = data.groupby([specialty_col, 'Time_Bands'])['Total'].mean().unstack()
        average_waiting_times[year] = grouped_data
    return average_waiting_times

# Analyzing average waiting times for inpatient and outpatient data
inpatient_average_waiting_times = analyze_average_waiting_times(inpatient_data)
outpatient_average_waiting_times = analyze_average_waiting_times(outpatient_data, is_outpatient=True)


print("Outpatient Average Waiting Times for 2018:")
print(outpatient_average_waiting_times[2018].head())  
```

```python
import matplotlib.pyplot as plt

def plot_waiting_times(data, specialty, title):
    plt.figure(figsize=(12, 6))
    for year, df in data.items():
        if specialty in df.index:
            plt.plot(df.columns, df.loc[specialty], label=f'Year {year}', marker='o')
    plt.title(title)
    plt.xlabel('Time Bands')
    plt.ylabel('Average Number of Patients')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example: Plotting for 'Cardiology' specialty in outpatient data
plot_waiting_times(outpatient_average_waiting_times, 'Cardiology', 'Outpatient Average Waiting Times for Cardiology (2018-2021)')

# Note: Repeat the plotting for other specialties or inpatient data by changing the parameters
```
